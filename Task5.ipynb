```python
import pandas as pd
# Load dataset
df = pd.read_excel("ApexPlanet_DataAnalytics_Dataset.xlsx")
# Remove missing values
df = df.dropna()
# Total Sales
total_sales = df["Total_Sales"].sum()
# Total Quantity
total_quantity = df["Quantity"].sum()
# Average Unit Price
average_unit_price = df["Unit_Price"].mean()
# Average Age
average_age = df["Age"].mean()
# Total Customers
total_customers = df["Customer_ID"].nunique()
# Total Orders
total_orders = df["Order_ID"].nunique()
# Sales by Category
sales_by_category = df.groupby("Category")["Total_Sales"].sum()
# Sales by City
sales_by_city = df.groupby("City")["Total_Sales"].sum()
# Sales by Gender
sales_by_gender = df.groupby("Gender")["Total_Sales"].sum()
# Quantity by Product
quantity_by_product = df.groupby("Product")["Quantity"].sum()
# Sales Trend
sales_trend = df.groupby("Order_Date")["Total_Sales"].sum()
# Display Results
print("Total Sales:", total_sales)
print("Total Quantity:", total_quantity)
print("Average Unit Price:", average_unit_price)
print("Average Age:", average_age)
print("Total Customers:", total_customers)
print("Total Orders:", total_orders)
print("\nSales by Category")
print(sales_by_category)
print("\nSales by City")
print(sales_by_city)
print("\nSales by Gender")
print(sales_by_gender)
print("\nQuantity by Product")
print(quantity_by_product)
print("\nSales Trend")
print(sales_trend)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_excel("ApexPlanet_DataAnalytics_Dataset.xlsx")
sales = df.groupby("Category")["Total_Sales"].sum()
plt.figure(figsize=(6,4))
plt.bar(sales.index, sales.values)
plt.title("Sales by Category")
plt.xlabel("Category")
plt.ylabel("Total Sales")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_excel("ApexPlanet_DataAnalytics_Dataset.xlsx")
sales = df.groupby("Category")["Total_Sales"].sum()
plt.figure(figsize=(4,4))
plt.pie(sales.values, labels=sales.index, autopct="%1.1f%%")
plt.title("Sales by Category")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_excel("ApexPlanet_DataAnalytics_Dataset.xlsx")
sales = df.groupby("Gender")["Total_Sales"].sum()
plt.figure(figsize=(4,4))
plt.pie(sales.values, labels=sales.index, autopct="%1.1f%%")
centre_circle = plt.Circle((0,0),0.70,color="white")
plt.gca().add_artist(centre_circle)
plt.title("Sales by Gender")
plt.show()
```
